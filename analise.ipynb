```python
# Verificando colunas com valores 0
# UTI_MES_IN	UTI_MES_AN	UTI_MES_AL	UTI_MES_TO	MARCA_UTI	UTI_INT_IN	UTI_INT_AN	UTI_INT_AL	UTI_INT_TO
colunas_com_zeros = (df == 0).sum()
colunas_com_zeros = colunas_com_zeros[colunas_com_zeros > 0]
colunas_com_zeros
```

## Tratamento dos nulos

```python
null_counts = df.isnull().sum()
colunas_com_nulos = null_counts[null_counts > 0]

print(colunas_com_nulos)
```

```python
threshold_para_remover_nulos = numero_linhas * 0.7
lista_colunas_para_remover = colunas_com_nulos[colunas_com_nulos > threshold_para_remover_nulos].index
lista_colunas_para_remover = lista_colunas_para_remover.union(colunas_com_zeros[colunas_com_zeros > threshold_para_remover_nulos].index)
lista_colunas_para_remover
```

```python
df_cleaned = df.drop(columns=lista_colunas_para_remover)
```

```python
# quantidade de colunas
print(f"Número de colunas após remoção: {len(df_cleaned.columns)}")
```

```python
#veja os valores da coluna COD_IDADE se existem valores diferentes de 4
df_cleaned['NACIONAL'].value_counts()
```

```python
df_cleaned.head(20)
```

```python
df_cleaned.to_csv('cleaned_data.csv', index=False)
```

# Separando códigos dos municípios

```python
df_munic = pd.read_csv('data/municipios.csv', encoding='ISO-8859-1')
```

```python
df_munic.head(5)
```

```python
df_munic = df_munic.iloc[3:].reset_index(drop=True)
df_munic.head(5)
```

```python
# 1. Cria as novas colunas 'codigo' e 'municipio' no DataFrame df_munic
#    extraindo os dados da coluna 'Indicadores Municipais'.
df_munic[['codigo', 'municipio']] = df_munic['Indicadores Municipais'].str.extract(r'^(\d+)\s(.*);.*$')

# 2. Remove a coluna original 'Indicadores Municipais' de df_munic,
#    pois os dados dela já foram separados.
df_munic = df_munic.drop(columns=['Indicadores Municipais'])

# Opcional: para visualizar o resultado
df_munic.head()
```

```python
df_munic.to_csv('municipios_tratado.csv', index=False)
```
